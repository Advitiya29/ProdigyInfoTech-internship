**TASK-1: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded file
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)

# Extract data for 2022
df_population = df[["Country Name", "2022"]].dropna()

# Get top 10 by population
top10_2022 = df_population.sort_values(by="2022", ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(top10_2022["Country Name"], top10_2022["2022"], color='skyblue')
plt.title("Top 10 Countries/Regions by Population in 2022")
plt.xlabel("Country/Region")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
